```python
import pandas as pd
data = pd.read_csv("Earthquakes_database.csv")

data = data.drop(range(2000, len(data)))

data.head()

```

```python
# 1 ( რომელიმე სვეტის მონაცემები გადაანაწილეთ რამდენიმე ახალ სვეტად, წაშალეთ რომელიმე სვეტი,)

data[['Hour', 'Minute', 'Second']] = data['Time'].str.split(':', expand=True)
data = data.drop(columns=['Time'])

data.head()
```

```python
# 2 (რიცხვითი ტიპის ველის მონაცემები დაყავით კატეგორიებად cut() ფუნქციის გამოყენებით (მაგ. ასაკის ველი 3 კატეგორიად „Children”, “Adolescent”, “Old”).)

data['Magnitude Level'] = pd.cut(data['Magnitude'], bins=[0, 3, 6, 10], labels=['Low', 'Medium', 'High'])
data.head()
```

```python
# 4 ააგეთ თქვენთვის სასურველი დიაგრამა
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(data['Magnitude Level'], data['Magnitude'])
```

```python
# 5 გამოიყენეთ თარიღებთან სამუშაო მინიმუმ 2 განსხვავებული ფუნქცია/ტრანსფორმაცია

data['Date'] = pd.to_datetime(data['Date'])

data['Day'] = data['Date'].dt.day
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year

data.head()
```
